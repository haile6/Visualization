```python
import pandas as pd
import plotly.express as px

# Sample project data
tasks = [
    {"Task": "Task A", "Start": "2025-05-01", "Finish": "2025-05-05"},
    {"Task": "Task B", "Start": "2025-05-02", "Finish": "2025-05-06"},
    {"Task": "Task C", "Start": "2025-05-04", "Finish": "2025-05-07"},
    {"Task": "Task D", "Start": "2025-05-06", "Finish": "2025-05-10"}
]

# Create DataFrame
df = pd.DataFrame(tasks)

# Create Gantt chart using Plotly Express
fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task", title="Project Timeline")

# Optional: Sort tasks from top to bottom by their start time
fig.update_yaxes(categoryorder="total ascending")

# Show the chart
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the DataFrame
data = {
    'Customer': [f'Customer {i+1}' for i in range(10)],
    'PurchaseAmount': [250, 600, 450, 800, 300, 120, 550, 700, 380, 900],
    'Profit': [80, 90, 200, 150, 120, 50, 95, 300, 100, 70]
}

df = pd.DataFrame(data)

# Step 2: Plot the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['PurchaseAmount'], df['Profit'], color='blue')

# Add labels
plt.title('Purchase Amount vs Profit')
plt.xlabel('Purchase Amount')
plt.ylabel('Profit')

# Step 3: Annotate customers with PurchaseAmount > 500 and Profit < 100
for i in range(len(df)):
    if df['PurchaseAmount'][i] > 500 and df['Profit'][i] < 100:
        plt.annotate(
            df['Customer'][i],
            (df['PurchaseAmount'][i], df['Profit'][i]),
            textcoords="offset points",
            xytext=(10, 13),
            ha='left',
            #color='red',
            arrowprops=dict(arrowstyle='->', color='red')
        )

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Simulated Dataset
data = pd.DataFrame({
    'ProductCategory': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Year': [2022]*6 + [2023]*6,
    'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb'],
    'Sales': [120, 150, 100, 200, 180, 130, 160, 170, 110, 220, 190, 140]
})

# 2. Pivot Table: Total sales per category by year and month
pivot = data.pivot_table(index=['Year', 'Month'], columns='ProductCategory', values='Sales', aggfunc='sum')

# Optional: Fill missing values with 0 (if any)
pivot = pivot.fillna(0)

# 3. Plot Stacked Bar Chart
pivot.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Total Sales by Product Category (Year & Month)')
plt.ylabel('Sales')
plt.xlabel('Year, Month')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```
